#### Misplaced values in "Processor" column
According to the query result, it looked like "Screen Size", "Color", "RAM Size", "SSD Capacity", "GPU", "Processor Speed" and "Brand" of the Dell laptops were all misplaced. I fixed the issue: 

```python
# I initiated this column so I could check whether the misplaced data were cleaned after I performed data manipulations. 
df_clean1['wrong_processor_value'] = None
```

```python
df_clean1.loc[df_clean1['Processor'] == 'dell', 'wrong_processor_value'] = 'dell'  # Used to track changes
df_clean1['Processor'] = np.where(df_clean1['wrong_processor_value'] == 'dell', df_clean1['Screen Size'], df_clean1['Processor'])
df_clean1['Screen Size'] = np.where(df_clean1['wrong_processor_value'] == 'dell', df_clean1['Color'], df_clean1['Screen Size'])
df_clean1['Color'] = np.where(df_clean1['wrong_processor_value'] == 'dell', df_clean1['RAM Size'], df_clean1['Color'])
df_clean1['RAM Size'] = np.where(df_clean1['wrong_processor_value'] == 'dell', df_clean1['SSD Capacity'], df_clean1['RAM Size'])
df_clean1['SSD Capacity'] = np.where(df_clean1['wrong_processor_value'] == 'dell', df_clean1['GPU'], df_clean1['SSD Capacity'])
df_clean1['GPU'] = np.where(df_clean1['wrong_processor_value'] == 'dell', df_clean1['Processor Speed'], df_clean1['GPU'])
df_clean1['Processor Speed'] = np.where(df_clean1['wrong_processor_value'] == 'dell', df_clean1['Brand'], df_clean1['Processor Speed'])
df_clean1.loc[df_clean1['wrong_processor_value'] == 'dell', 'Brand'] = 'dell'
```

```python
def clean_screen_size(text):
    text = text.replace('not available', '0')
    text = text.replace('Does not apply', '0')
    text = text.replace('missing', '0')
    text = text.replace('nan', '0')
    text = text.replace('None', '0')
    text = text.replace('IPS1920x1080(FullHD', '') 
    text = text.replace('QHD+ Touch', '')   
    text = text.replace('and ampquot in', '')
    text = text.replace('touchscreen', '')
    text = text.replace('IPS touchscreen 1920 x 1080 (Full HD)', '')
    text = text.replace(' in FHD', '')
    text = text.replace('FHD', '')
    text = text.replace('-inch', '')
    text = text.replace('TOUCHSCREEN', '')
    text = text.replace('in.', '')
    text = text.replace('"', '')
    text = text.replace(' ', '')
    text = text.replace('in', '')
    text = text.replace('”', '')
    return text
```

```python
def clean_color(text):
    text = str(text)
    text = text.lower()
    text = text.replace('nan', 'missing')
    text = text.replace('light blue', 'blue')
    text = text.replace('light gray', 'gray')
    text = text.replace('dark gray', 'gray')
    text = text.replace('epic gray', 'gray')
    text = text.replace('black/gray', 'multicolor')
    text = text.replace('Silver/Black', 'multicolor')
    text = text.replace('black / blue', 'multicolor')
    text = text.replace('silver/black', 'multicolor')
    text = text.replace('universe blue', 'blue')
    text = text.replace('eclipse gray', 'gray')
    text = text.replace('grey', 'gray')
    return text
```

```python
def clean_RAM(text):
    text = str(text)
    text = text.lower()
    text = text.replace('na', 'nan')
    text = text.replace('nann', 'nan')
    text = text.replace('16384mb (16gb)', '16')
    text = text.replace(' ', '')
    text = text.replace('gb', '')
    text = text.replace('ddr4', '')
    text = text.replace('ddr3', '')
    text = text.replace('mb', '')
    text = text.replace('16384', '16')
    text = text.replace('(2x4)', '')
    text = text.replace('ssd', '')
    text = text.replace('noram', '0')
    return text
```

```python
def clean_processor_speed(text):
    text = str(text)
    text = text.replace('na', 'nan')
    text = text.replace('missing', 'nan')
    text = text.replace('nann', 'nan')
    text = text.replace('2.50 - 2.99 ghz', '2.5')
    text = text.replace('2.50 - 2.99 GHz', '2.5')
    text = text.replace('QUAD 2.90 GHz - 3.80 GHz Turbo Boost', '2.9')
    text = text.replace('1.8Ghz With TurboBoost to 4.9Ghz', '1.8')
    text = text.replace('1.50 - 1.99 ghz', '1.5')
    text = text.replace('base 2.6GHz, up to 4.4GHz', '2.6')
    text = text.lower()
    text = text.replace('ghz', '')
    text = text.replace(' ', '')
    return text
```

```python
df_clean1['Processor Speed'] = df_clean1['Processor Speed'].apply(lambda x: clean_processor_speed(x))
df_clean1['Processor Speed'] = df_clean1['Processor Speed'].astype('float')
```

```python
def clean_hdd_capacity(text):
    text = str(text)
    text = text.replace('na', 'nan')
    text = text.replace('Not Applicable', 'nan')
    text = text.replace('NO HDD', '0')
    text = text.replace('does not apply', 'nan')
    text = text.replace('nann', 'nan')
    text = text.replace('Not Included', '0')
    text = text.replace(' or more', '2')
    text = text.replace('2 TB', '2000')
    text = text.replace('1 TB', '1000')
    text = text.lower()
    text = text.replace('gb', '')
    text = text.replace('g', '')
    text = text.replace('no', '0')
    text = text.replace(' ', '')
    return text
```

```python
def clean_storage(text):
    text = text.replace('not included', 'no drive')
    text = text.replace('ssd (solid state drive)', 'ssd')
    text = text.replace('ssd solid state drive', 'ssd')
    text = text.replace('1tb ssd storage + 32gb intel optane cache', 'ssd')
    text = text.replace('m.2 ssd', 'ssd')
    text = text.replace('m.2', 'ssd')
    text = text.replace('no hdd', 'no drive')
    text = text.replace('pcie nvme class 35 solid state drive', 'ssd')
    text = text.replace('nvme m.2 ssd', 'ssd')
    text = text.replace('na', 'missing')
    text = text.replace('pcie nvme class 35 2 230 solid state drive', 'ssd')
    text = text.replace('hdd (hard disk drive)', 'hdd')
    text = text.replace('nvme ssd ssd', 'ssd')
    text = text.replace('other', 'missing')
    text = text.replace('ssd ssd', 'ssd')
    return text
df_clean1['Storage Type'] = df_clean1['Storage Type'].apply(lambda x: clean_storage(x))
df_clean1['Storage Type'].unique()
```
